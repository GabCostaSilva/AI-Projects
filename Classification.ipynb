# Classificadores

## Dataset Divorce

* 170 exemplos
* 54 atributos
* 2 classes

```python
import pandas as pd
import numpy as np
from classifier import Classifier



dataset = pd.read_csv("./datasets/divorce.csv", skiprows=1, header=None, sep=";")

dataset_np = np.array(dataset)

data = dataset_np[:, :-1]
target = dataset_np[:, -1]

clf = Classifier(data, target, data[:, -1].shape)

knn = {
    'n_neighbors': [1],
    'metric': ['euclidean'],
    'weights': ['uniform']
}
tree = {
    'max_depth': [3]
}
regression = {
    "solver": ["lbfgs"],
    "penalty": ["l2"],
    "dual": [False],
    "tol": [1e-4],
    "C": [1.],
    "max_iter": [100]
}
mlp = {
    "hidden_layer_sizes": [(100,)],
    "activation": ['relu'],
    "solver": ["lbfgs"],
    "max_iter": [200]
}


clf.run(knn_params=knn,
            decision_tree_params=tree,
            regression_params=regression,
            mlp_params=mlp, nb_params={}
            )
```

## Resultados

```python
clf.graph_plot()
```

# Dataset Contraceptive Methods
  Este conjunto de dados é um subconjunto do Contraceptivo Nacional da Indonésia de 1987.
   Pesquisa de Prevalência. As amostras são de mulheres casadas que ou não eram 
   grávidas ou não sabem se estavam no momento da entrevista. O 
   O problema é prever a escolha do método contraceptivo atual 
   (sem uso, métodos de longo prazo ou de curto prazo) de uma mulher baseada 
   sobre suas características demográficas e sócio-econômicas.

* 1473 exemplos
* 9 atributos
* 3 classes

```python
from classifier import Classifier
contraceptive = pd.read_csv("./datasets/contraceptive_methods.data", sep=",")
contraceptive_np = np.array(contraceptive)
data = contraceptive_np[:, :-1]
target = contraceptive_np[:, -1]
```

```python
clf = Classifier(data, target, target.shape)

knn = {
    'n_neighbors': [9],
    'metric': ['euclidean'],
    'weights': ['uniform']
}
tree = {
    'max_depth': [5],
    "criterion": ["gini"]
}
regression = {
    "solver": ["lbfgs"],
    "penalty": ["none"],
    "dual": [False],
    "tol": [1e-4],
    "C": [1],
    "max_iter": [200]
}
mlp = {'activation': ['relu'], 'hidden_layer_sizes': [(100,)], 'max_iter': [500], 'solver': ['adam']}

nb = {
    "var_smoothing": [1e-9, 1e-4, 1e15]
}

clf.run(knn_params=knn, decision_tree_params=tree,
            regression_params=regression,
            mlp_params=mlp, nb_params=nb
            )
```

```python
clf.graph_plot()
```

# Dataset ecoli
* 336 exemplos
* 8 atributos
* 9 classes
* 143 exemplos classe majoritaria
1.  Nome seqüencial: Número de adesão para o banco de dados SWISS-PROT
2. mcg: Método de McGeoch para reconhecimento da seqüência de sinais.
3. gvh: método de von Heijne para reconhecimento da seqüência de sinais.
4. lip: pontuação consensual da seqüência de sinais Peptidase II de von Heijne. Atributo binário.
5. chg: Presença de carga em N-terminus de lipoproteínas previstas. Atributo binário.
6. aac: escore de análise discriminante do conteúdo de aminoácidos de membrana externa e proteínas periplásmicas.
7. alm1: pontuação do programa de previsão da região de extensão da membrana ALOM.
8. alm2: escore do programa ALOM depois de excluir sinal putativo de clivagem. regiões da seqüência.
